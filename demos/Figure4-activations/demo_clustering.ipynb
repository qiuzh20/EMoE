```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# Load the provided data
```

```python
selected_data = np.load('selected_data_2_64.npy')
sorted_data = np.load('sorted_data_2_64.npy')
def visualize_3d_fully_aligned_axes(matrix1, title1, matrix2, title2):
    fig = plt.figure(figsize=(24, 8))
    
    # Set common axes limits for alignment
    xlims = (0, 3072)
    ylims = (0, 320)
    zlims = (np.min([matrix1, matrix2]), np.max([matrix1, matrix2]))
    
    # First subplot for matrix1
    ax1 = fig.add_subplot(121, projection='3d')
    y1, x1 = np.where(matrix1 != 0)
    z1 = matrix1[y1, x1]
    ax1.scatter(x1, y1, z1, s=10, c=z1, cmap="Blues", marker='o', alpha=0.5)
    ax1.set_xlim(xlims)
    ax1.set_ylim(ylims)
    ax1.set_zlim(zlims)
    ax1.set_xlabel('Neuron Index', fontsize=17)
    ax1.set_ylabel('Sample Index', fontsize=17)
    ax1.view_init(elev=30, azim=-60)
    ax1.set_title(title1, fontsize=20, y=1)
    ax1.tick_params(axis='both', which='major', labelsize=12)
    
    # Second subplot for matrix2
    ax2 = fig.add_subplot(122, projection='3d')
    y2, x2 = np.where(matrix2 != 0)
    z2 = matrix2[y2, x2]
    ax2.scatter(x2, y2, z2, s=10, c=z2, cmap="Blues", marker='o', alpha=0.5)
    ax2.set_xlim(xlims)
    ax2.set_ylim(ylims)
    ax2.set_zlim(zlims)
    ax2.set_xlabel('Neuron Index', fontsize=17)
    ax2.set_ylabel('Sample Index', fontsize=17)
    ax2.view_init(elev=30, azim=-60)
    ax2.set_title(title2, fontsize=20, y=1)
    ax2.tick_params(axis='both', which='major', labelsize=12)
    
    plt.subplots_adjust(wspace=-0.5)
    plt.savefig('Cluster_results.pdf', bbox_inches='tight', pad_inches=0.0)
    plt.show()

# Visualize the two matrices with fully aligned axes
visualize_3d_fully_aligned_axes(selected_data, "Raw Data Activation Patterns",
                                sorted_data, "Weight-Clustered Activation Patterns")
```

```python
selected_data = np.load('selected_data_2_64.npy')
sorted_data = np.load('sorted_data_2_64.npy')
xlims = (0, 3072)
ylims = (0, 320)
# Define a function to visualize the data in 3D scatter plot with adjusted axes
def visualize_3d_axes_adjusted(matrix, title):
    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    # Get the y, x, z indices of non-zero values to adjust axes
    y, x = np.where(matrix != 0)
    z = matrix[y, x]
    
    ax.scatter(x, y, z, s=10, c=z, cmap="Blues", marker='o', alpha=0.5)
    ax.set_xlim(xlims)
    ax.set_ylim(ylims)
    ax.set_xlabel('Neuron Index', fontsize=14)
    ax.set_ylabel('Sample Index', fontsize=14)
    ax.view_init(elev=30, azim=-60)  # Adjust the viewing angle for better visualization
    ax.set_title(title, fontsize=14, y=1)
    ax.tick_params(axis='both', which='major', labelsize=12)
    plt.savefig(f'{title}.pdf', bbox_inches='tight', pad_inches=0.0)
    plt.show()

# Create adjusted 3D scatter plots for both matrices with axes adjustments
visualize_3d_axes_adjusted(selected_data, "Raw Data Activation Patterns")
visualize_3d_axes_adjusted(sorted_data, "Weight-Clustered Activation Patterns")
```
