```python
# Provide the code for the user
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = [
    [1, 0.74],
    [0.9, 0.52],
    [0.8, 0.45],
    [0.7, 0.57],
    [0.6, 0.62],
    [0.5, 0.36],
    [0.3, 0.44],
    [0.1, 0.47]
]
data = np.array(data).T
```

```python
EMoE = [84.49, 84.27, 83.83, 83.13, 82.43, 81.45, 78.67, 72.15][:-2][::-1]
Dense = [83.75, 83.42, 83.05, 82.56, 81.81, 80.91, 78.23, 71.68][:-2][::-1]

fig = plt.figure(figsize=(6, 4))
x = data[0][:-2]  # Data Ratio
plt.plot(x, EMoE, marker='o', linestyle='-', label="EMoE")
plt.plot(x, Dense, marker='o', linestyle='--', label="Dense")

plt.axhline(y=83.75, color='y', linestyle='-.', linewidth=1.0, label='Dense All')

plt.xlabel('Data Ratio', fontsize=18)
plt.ylabel('Accuracy', fontsize=18)
fig.axes[0].set_xticklabels(['0', '1.0', '0.9', '0.8', '0.7', '0.6', '0.5'], fontsize=16)
plt.ylim(80.5, 84.8)
plt.yticks(np.arange(81, 85, 1))
plt.yticks(fontsize=14)
plt.grid(True)
plt.legend(fontsize=18)
plt.tight_layout()
plt.savefig('ratio_compare_acc.png', bbox_inches='tight', pad_inches=0.0)
plt.savefig('ratio_compare_acc.pdf', bbox_inches='tight', pad_inches=0.0)
```
